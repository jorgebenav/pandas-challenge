```python
import pandas as pd
from pathlib import Path
```

```python
#read and print the school and student information 
school_raw = Path("Resources/schools_complete.csv")
school_read = pd.read_csv(school_raw)
school_read.rename(columns={'school_name': 'School Name', 'size': 'Total students population', 'budget':'Total school budget'}, inplace=True)

student_raw = Path("Resources/students_complete.csv")
student_read = pd.read_csv(student_raw)
student_read.rename(columns={'student_name': 'Students Name','school_name': 'School Name','reading_score': 'Reading Score', 'math_score': 'Math Score'}, inplace=True)
school_read
```

```python
#District Summary# The following object will be achieved
#in the new dateframe
#first-Total # of unique schools
district_school_number = len(school_read['School Name'].unique())
district_school = pd.DataFrame({'Number of Schools': [district_school_number]})

#Total # students
district_school['Total students'] = school_read['Total students population'].sum()


#Total budget
district_school['Total budget'] = school_read['Total school budget'].sum()
#Average math score 
#for this we will use the student_read dataset rather the school_read
#using round(), to make it pretty 
district_school['Average Math Score'] = round(student_read['Math Score'].mean(), 4)
#Average reading score
district_school['Average Reading Score'] = round(student_read['Reading Score'].mean(), 4)
#% passing math (the percentage of students who passed math)
passing_math_count = len(student_read[student_read['Math Score'] >= 70])
district_school['% passing math'] = round((passing_math_count / district_school['Total students'].astype(float)) * 100, 4)
#% passing reading (the percentage of students who passed reading)
passing_reading_count = len(student_read[student_read['Reading Score'] >= 70])
district_school['% passing read'] = round((passing_reading_count / district_school['Total students'].astype(float)) * 100, 4)
#% overall passing (the percentage of students who passed math AND reading)
passing_both = len(student_read[(student_read['Reading Score'] >= 70) & (student_read['Math Score'] >= 70)])
district_school['% passing both'] = round((passing_both / district_school['Total students'].astype(float)) * 100, 4)

#print dateframe
district_school
```

```python
#School Summary# The following object will be achieved
#in the new dateframe, we can use the  
#School name
#School type
#Total students
#Total school budget
school_summary = school_read.set_index("School Name")[["type", "Total students population", "Total school budget"]]


# Sorting the DataFrame by the index (School Name) alphabetically
school_summary = school_summary.sort_index()

school_summary['Per student budget'] = round(school_summary['Total school budget'] / school_summary['Total students population'], 4)


# Calculating the average math score
school_summary['Average math score'] = round(student_read.groupby('School Name')['Math Score'].mean(), 4)

# Calculating the average reading score
school_summary['Average Reading score'] = round(student_read.groupby('School Name')['Reading Score'].mean(), 4)

# Calculating the percentage of students passing math
passing_math = student_read[student_read['Math Score'] >= 70]
passing_m = passing_math.groupby('School Name').size()
total_population = school_summary["Total students population"]
school_summary['passing math'] = round((passing_m / total_population) * 100, 4)

# Calculating the percentage of students passing reading
passing_reading = student_read[student_read['Reading Score'] >= 70]
passing_r = passing_reading.groupby('School Name').size()
school_summary['passing reading'] = round((passing_r / total_population) * 100, 4)

# Calculating the percentage of students passing both math and reading
passing_both = student_read[(student_read['Math Score'] >= 70) & (student_read['Reading Score'] >= 70)]
passing_both_counts = passing_both.groupby('School Name').size()
school_summary['Passing Both'] = round((passing_both_counts / total_population) * 100, 4)

# Displaying the final DataFrame
school_summary
```

```python
top_school_raw = school_summary.sort_values(by='Passing Both', ascending=False)
top_school =top_school_raw.head(5)
top_school 
```

```python
#MathScores by Grade
math_grade = pd.DataFrame(school_read["School Name"])
ninth_math = student_read[student_read["grade"] == "9th"]
tenth_math = student_read[student_read["grade"] == "10th"]
eleventh_math = student_read[student_read["grade"] == "11th"]
twelfth_math = student_read[student_read["grade"] == "12th"]

# average math scores by school for each grade
ninth_grade_avg = round(ninth_math.groupby('School Name')['Math Score'].mean().rename('9th'), 2)
tenth_grade_avg = round(tenth_math.groupby('School Name')['Math Score'].mean().rename('10th'), 2)
eleventh_grade_avg = round(eleventh_math.groupby('School Name')['Math Score'].mean().rename('11th'), 2)
twelfth_grade_avg = round(twelfth_math.groupby('School Name')['Math Score'].mean().rename('12th'), 2)

# add the columns
math_grade = pd.concat([ninth_grade_avg, tenth_grade_avg, eleventh_grade_avg, twelfth_grade_avg], axis=1)

# Display the resulting DataFrame
math_grade
```

```python
reading_grade = pd.DataFrame(school_read["School Name"])
ninth_read = student_read[student_read["grade"] == "9th"]
tenth_read = student_read[student_read["grade"] == "10th"]
eleventh_read = student_read[student_read["grade"] == "11th"]
twelfth_read = student_read[student_read["grade"] == "12th"]

# average math scores by school for each grade
ninth_read = round(ninth_read.groupby('School Name')['Reading Score'].mean().rename('9th'), 2)
tenth_read = round(tenth_read.groupby('School Name')['Reading Score'].mean().rename('10th'), 2)
eleventh_read = round(eleventh_read.groupby('School Name')['Reading Score'].mean().rename('11th'), 2)
twelfth_read = round(twelfth_read.groupby('School Name')['Reading Score'].mean().rename('12th'), 2)

# add the columns
read_grade = pd.concat([ninth_read, tenth_read, eleventh_read, twelfth_read], axis=1)

# Display the resulting DataFrame
read_grade
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
school_summary['Average Spending Ranges'] = pd.cut(school_summary['Per student budget'], bins=spending_bins, labels=labels, include_lowest=True)
school_summary
```

```python
metrics = ['Average math score', 'Average Reading score', 'passing math', 'passing reading', 'Passing Both']
spending_summary = school_summary.groupby('Average Spending Ranges', observed=True)[metrics].mean()
spending_summary
```

```python
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# appending a new column 
school_summary['School Size'] = pd.cut(school_summary['Total students population'], bins=size_bins, labels=size_labels, include_lowest=True)
school_summary
```

```python
sizemetrics = ['Average math score', 'Average Reading score', 'passing math', 'passing reading', 'Passing Both']
population_summary = school_summary.groupby('School Size', observed=True)[sizemetrics].mean()
population_summary
```
